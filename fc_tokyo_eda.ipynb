```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Hiragino Sans'
```

```python
df = pd.read_csv("J1_tokyo_home_2015-2024.csv")
```

```python
# 年別入場者数可視化(棒グラフ)

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="シーズン", y="入場者数", ci=None)
plt.title("FC東京 ホームゲーム 年別入場者数 (2015-2024)")
plt.xlabel("シーズン")
plt.ylabel("入場者数")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 年別入場者数可視化(箱ひげ図)
plt.rcParams['font.family'] = 'Hiragino Sans'
plt.figure(figsize=(10, 6))
sns.boxplot(x="シーズン", y="入場者数", data=df)
plt.title("FC東京ホームゲームの年別入場者数分布 (2015-2024)")
plt.xlabel("シーズン")
plt.ylabel("入場者数")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 2015年の入場者数推移(ヒストグラム)
df_2015 = df[df["シーズン"] == 2015]
plt.figure(figsize=(10, 6))
sns.histplot(data=df_2015, x="入場者数", bins=10)
plt.title("2015年の入場者数推移(ヒストグラム)")
plt.xlabel("入場者数")
plt.ylabel("試合数")
```

```python
# 2024年の入場者数推移(ヒストグラム)
df_2015 = df[df["シーズン"] == 2024]
plt.figure(figsize=(10, 6))
sns.histplot(data=df_2015, x="入場者数", bins=10)
plt.title("2024年の入場者数推移(ヒストグラム)")
plt.xlabel("入場者数")
plt.ylabel("試合数")
```

```python
# 2015年の入場者数推移(折れ線グラフ)
df_2015 = df[df["シーズン"] == 2015]
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_2015, x="試合日", y="入場者数", marker="o")
plt.title("FC東京 2015年ホームゲームの入場者数推移")
plt.xlabel("日付")
plt.ylabel("入場者数")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 2019年の入場者数推移(折れ線グラフ)
df_2019 = df[df["シーズン"] == 2019]
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_2019, x="試合日", y="入場者数", marker="o")
plt.title("FC東京 2019年ホームゲームの入場者数推移")
plt.xlabel("日付")
plt.ylabel("入場者数")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 2020年の入場者数推移(折れ線グラフ)
df_2024 = df[df["シーズン"] == 2024]
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_2024, x="試合日", y="入場者数", marker="o")
plt.title("FC東京 2024年ホームゲームの入場者数推移")
plt.xlabel("日付")
plt.ylabel("入場者数")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# df_2024のスタジアムが味スタのみ抽出
df_2024_ajisuta = df_2024[df_2024["スタジアム"] == "味スタ"]
df_2024_ajisuta
```

```python
len(df_2024_ajisuta)
```

```python
df_2019 = df[df["シーズン"] == 2019]
df_2019
```

```python
len(df_2019)
```

```python
# df_2024_ajisuta と df_2019 の入場者数を比較(箱ひげ図)
combined_df = pd.concat([df_2019.assign(Year='2019'), df_2024_ajisuta.assign(Year='2024')])
plt.figure(figsize=(8, 6))
sns.boxplot(x='Year', y='入場者数', data=combined_df)
plt.title('FC東京 ホームゲーム 入場者数比較 (2019年 vs 2024年国立除く)')
plt.xlabel('年')
plt.ylabel('入場者数')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# df_2024 と df_2019 の入場者数を比較(箱ひげ図)
combined_df = pd.concat([df_2019.assign(Year='2019'), df_2024.assign(Year='2024')])
plt.figure(figsize=(8, 6))
sns.boxplot(x='Year', y='入場者数', data=combined_df)
plt.title('FC東京 ホームゲーム 入場者数比較 (2019年 vs 2024年国立含む)')
plt.xlabel('年')
plt.ylabel('入場者数')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
