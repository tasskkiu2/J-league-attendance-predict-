# J1 Data Scraping

J.League Data Site から年度別データを取得し、FC東京ホームゲームのCSVを作成する。


```python
from pathlib import Path

PROJECT_ROOT = Path.cwd()
while (
    PROJECT_ROOT != PROJECT_ROOT.parent
    and not (PROJECT_ROOT / "requirements.txt").exists()
):
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_DATA_DIR = DATA_DIR / "raw"
PROCESSED_DATA_DIR = DATA_DIR / "processed"
```

```python
import pandas as pd
import requests
```

```python
url = "https://data.j-league.or.jp/SFMS01/search?competition_years=2025&competition_frame_ids=1&competition_ids=589&tv_relay_station_name="
headers = {"User-Agent": "Mozilla/5.0"}
resp = requests.get(url, headers=headers, timeout=10)
resp.raise_for_status()

tables = pd.read_html(resp.text)
df_2024 = tables[0]
print(df_2024.head())
```
