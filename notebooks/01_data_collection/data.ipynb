# Attendance Data Exploration

2024年のJ1入場者データをチーム別に確認する。


```python
from pathlib import Path

PROJECT_ROOT = Path.cwd()
while (
    PROJECT_ROOT != PROJECT_ROOT.parent
    and not (PROJECT_ROOT / "requirements.txt").exists()
):
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_DATA_DIR = DATA_DIR / "raw"
PROCESSED_DATA_DIR = DATA_DIR / "processed"
```

```python
import pandas as pd
```

```python
import pandas as pd
import requests

url = "https://data.j-league.or.jp/SFMS01/search?competition_years=2024&competition_frame_ids=1&competition_ids=589&tv_relay_station_name="
headers = {"User-Agent": "Mozilla/5.0"}
resp = requests.get(url, headers=headers, timeout=10)
resp.raise_for_status()

tables = pd.read_html(resp.text)
df = tables[0]
print(df.head())
```

```python
df
```

```python
# ホーム:広島のカラムのみ抜き出し
df_hiroshima = df[df["ホーム"] == "広島"]
df_hiroshima
```

```python
# 試合それぞれの観客数を可視化
# 日本語フォントの設定

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
plt.rcParams["font.family"] = "Hiragino Sans"
sns.barplot(data=df_hiroshima, x="節", y="入場者数")
plt.title("広島ホームゲームの観客数（2024年J1リーグ）")
plt.xlabel("節")
plt.ylabel("観客数")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ホーム:FC東京のカラムのみ抜き出し､可視化
df_fctokyo = df[df["ホーム"] == "FC東京"]
plt.figure(figsize=(10, 6))
plt.rcParams["font.family"] = "Hiragino Sans"
sns.barplot(data=df_fctokyo, x="節", y="入場者数")
plt.title("FC東京ホームゲームの観客数（2024年J1リーグ）")
plt.xlabel("節")
plt.ylabel("観客数")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# fc東京の味スタの観客数を可視化
df_fctokyo = df[(df["ホーム"] == "FC東京") & (df["スタジアム"] == "味スタ")]
plt.figure(figsize=(10, 6))
plt.rcParams["font.family"] = "Hiragino Sans"
sns.barplot(data=df_fctokyo, x="節", y="入場者数")
plt.title("FC東京ホーム味スタの観客数（2024年J1リーグ）")
plt.xlabel("節")
plt.ylabel("観客数")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# アウェイ:FC東京の カラムのみ抜き出し､可視化
df_fctokyo_away = df[df["アウェイ"] == "FC東京"]
plt.figure(figsize=(10, 6))
plt.rcParams["font.family"] = "Hiragino Sans"
sns.barplot(data=df_fctokyo_away, x="節", y="入場者数")
plt.title("FC東京アウェイゲームの観客数（2024年J1リーグ）")
plt.xlabel("節")
plt.ylabel("観客数")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
